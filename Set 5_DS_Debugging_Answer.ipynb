```python
import pandas as pd
from scipy.stats import ttest_ind

data = pd.DataFrame({'Group': ['A', 'A', 'B', 'B'], 'Score': [70, 80, 85, 90]})

group_A = data[data['Group'] == 'A']['Score']
group_B = data[data['Group'] == 'B']['Score']

t_stat, p_value = ttest_ind(group_A, group_B, axis=1)
print(f"T-Statistic: {t_stat}, P-Value: {p_value}")
```

```python
import pandas as pd
from scipy.stats import f_oneway

data = pd.DataFrame({
    'Course': ['Math', 'Math', 'Science', 'Science', 'English', 'English', 'Math', 'Science', 'English'],
    'Marks': [65, 70, 85, 80, 60, 75, 72, 90, 68]
})

f_stat, p_value = f_oneway(data['Marks'])
print("F-Statistic:", f_stat)
print("P-Value:", p_value)
```

```python
#Answer: Here we try to perform analysis of variance using f_oneway test. but wrongly mentioned marks instead of group of course. then only it can find the varaince between the groups.
import pandas as pd
from scipy.stats import f_oneway

data = pd.DataFrame({
    'Course': ['Math', 'Math', 'Science', 'Science', 'English', 'English', 'Math', 'Science', 'English'],
    'Marks': [65, 70, 85, 80, 60, 75, 72, 90, 68]
})

math = data[data['Course']=='Math']['Marks']
science = data[data['Course']=='Science']['Marks']
english = data[data['Course']=='English']['Marks']
f_stat, p_value = f_oneway(math,science,english)
print("F-Statistic:", f_stat)
print("P-Value:", p_value)
```
