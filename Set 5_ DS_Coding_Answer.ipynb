## Question 1:A company wants to analyze whether there is a significant difference in the average sales across four regions using One-Way ANOVA.

```python
import pandas as pd
from scipy.stats import f_oneway

data = pd.DataFrame({'Region': ['North', 'North', 'South', 'South', 'East', 'East', 'West', 'West'],
    'Sales': [200, 220, 250, 240, 180, 210, 230, 220]})

group1= data[data['Region'] == 'North']['Sales']
group2= data[data['Region'] == 'South']['Sales']
group3= data[data['Region'] == 'East']['Sales']
group4= data[data['Region'] == 'West']['Sales']

f_test,p_value = f_oneway(group1,group2,group3,group4)

print(f"F-statistics:{f_test}")
print(f"p_value:{p_value}")
```

## Question 2:Analyze whether the employee performance scores are affected by the type of training and experience using Two-Way ANOVA.

```python
import pandas as pd
import statsmodels.api as sm
from statsmodels.formula.api import ols
data = pd.DataFrame({'Training': ['A', 'A', 'B', 'B', 'A', 'B', 'A', 'B'],
    'Experience': ['Low', 'High', 'Low', 'High', 'Low', 'High', 'Low', 'High'],
    'Performance': [75, 80, 85, 90, 78, 88, 82, 89]})
model = ols('Performance ~ C(Training) * C(Experience)', data=data).fit()
anova_table = sm.stats.anova_lm(model, typ=2)

anova_table
```

## Question 3: A real estate company wants to check for multicollinearity using VIF to predict house prices.

```python
import pandas as pd
from statsmodels.stats.outliers_influence import variance_inflation_factor

data = pd.DataFrame({'Area': [1500, 2000, 2500, 3000, 3500],
    'Bedrooms': [3, 4, 3, 5, 4],
    'Bathrooms': [2, 3, 2, 4, 3],
    'House_Price': [300000, 400000, 350000, 500000, 450000]})

Feature = data[['Area','Bedrooms','Bathrooms','House_Price']]

vif_data = pd.DataFrame()
vif_data["Feature"] = Feature.columns
vif_data["VIF"] = [variance_inflation_factor(Feature.values,i) for i in range(len(Feature.columns))]
print(vif_data)
```

## Question 4:A school tested two teaching methods (Traditional vs Interactive). Perform an Independent T-Test to see if their average scores are different.

```python
import pandas as pd
from scipy.stats import ttest_ind

data= pd.DataFrame({'Method': ['Traditional', 'Traditional', 'Interactive', 'Interactive'],
    'Score': [70, 75, 85, 90]})

traditional_grp= data[data['Method']=='Traditional']['Score']
interactive_grp= data[data['Method']== 'Interactive']['Score']

t_stat,p_value = ttest_ind(traditional_grp,interactive_grp,axis=0)

print(f"Statistic:{t_stat} \nP-Value: {p_value}")
```

## Question 5: A fitness center wants to analyze if workout time is positively correlated with calorie burn using Pearson correlation.


```python
import pandas as pd
from scipy.stats import pearsonr
data = pd.DataFrame({
    'Workout_Time': [30, 45, 60, 90, 120],
    'Calories_Burned': [200, 300, 400, 500, 600]
})

correlation,p_value = pearsonr(data['Workout_Time'],data['Calories_Burned'])

print(f"Correlation Coefficient:{correlation} \nP-value:{p_value}")

```
