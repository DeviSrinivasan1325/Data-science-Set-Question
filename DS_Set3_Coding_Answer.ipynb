## Question 1: A company wants to analyze customer satisfaction survey data. How can Data Science concepts help to clean and analyze this data?

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

#Sample data
data= pd.DataFrame({ 'CustomerID': [1, 2, 3, 4, 5, 5, 6, 7, 8, 9],
    'Satisfaction': [8, 7, np.nan, 9, 6, 6, 10, 4, 5, 7]})

#Data preprocessing
#handling missing value using fillna
data['Satisfaction'].fillna(data['Satisfaction'].median(),inplace=True)

#Remove duplicate entries using Customer ID
data.drop_duplicates(subset='CustomerID', keep='first',inplace=True)

#Satistical summary
print("Satistical Summary")
print(data['Satisfaction'].describe())

#Data visualization
sns.histplot(data['Satisfaction'],kde=True)
plt.title('Customer Satisfaction Distribution')
plt.xlabel('Satisfaction score')
plt.ylabel('Frequency')
plt.show()

#box plot for otliers
sns.boxplot(x=data['Satisfaction'])
plt.title('Customer Satisfaction Distribution using box plot')
plt.xlabel('Satisfaction score')
plt.ylabel('Frequency')
plt.show()
```

## Question 2:  A retail store has recorded monthly sales data for its products. How can Data Science concepts be applied to understand sales performance?

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

dataset= pd.DataFrame({'Month': ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun'],
 'Sales': [1500, 1700, 1600, 1800, 2200, 2100]})


print("The Satistical Summary of Sales:")
print(dataset['Sales'].describe())


#converting month as categorical data, otherwise order will be messed.graph will be disaster.
month_order= ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun']
dataset['Month'] = pd.Categorical(dataset['Month'], categories= month_order, ordered=True)

#Line plot for showing the sales trend
sns.lineplot(x=dataset['Month'], y=dataset['Sales'],marker='o', color='blue')
plt.title("Monthly Sales trend")
plt.xlabel("Month")
plt.ylabel('Sales')
plt.grid(True)
plt.show()

#Bar plot for sales comparison.
sns.barplot(x=dataset['Month'], y=dataset['Sales'],palette='viridis')
plt.title("Monthly Sales comparison")
plt.xlabel("Month")
plt.ylabel('Sales')
plt.show()
```

## Question 3:A hospital conducted a study to check the effectiveness of a new medical test. How can Data Science concepts be used to validate its results?

```python
from scipy import stats 

group_a = [110, 115, 108, 102, 107]
group_b = [120, 125, 130, 122, 128]

t_test,p_value = stats.ttest_ind(group_a,group_b)
print(f'T-Statistic: {t_test:.2f}, P-Value:{p_value:.4f}')

if (p_value <=0.05):
    print('Reject Null Hypothesis: Test is effective')
else:
    print('Fails to Reject Null Hypothesis: Test is not effective')
```

## Question 4:  A company wants to evaluate employee performance using various metrics. How can Data Science concepts be applied?

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

dataset= pd.DataFrame({ 'Department': ['HR', 'IT', 'Sales', 'IT', 'HR', 'Sales', np.nan, 'HR', 'Sales', 'IT'],
    'Experience (Years)': [1.0, 3.0, 5.0, np.nan, 10.0, 2.0, 8.0, 4.0, np.nan, 9.0],
    'Projects Completed': [2.0, 5.0, np.nan, 10.0, 12.0, 3.0, 9.0, np.nan, 14.0, 11.0],
    'Performance Rating': [3.0, np.nan, 7.0, 8.0, 9.0, 4.0, 8.0, 7.0, 10.0, np.nan]})
print("Original data with missing values:\n",dataset)
```

```python
#data cleanizing
# handling missing value
dataset['Experience (Years)'].fillna( dataset['Experience (Years)'].median(), inplace=True)
dataset['Projects Completed'].fillna( dataset['Projects Completed'].median(), inplace=True)
dataset['Performance Rating'].fillna( dataset['Performance Rating'].median(), inplace=True)
dataset['Department'].fillna(dataset['Department'].mode()[0], inplace=True)
print("Cleaned data:\n",dataset)
```

```python
#Descriptive statistics:
print("Descriptive statistics:\n",dataset.describe())
```

```python
corr_matrix = dataset.select_dtypes(include= ['float64','int64']).corr()
#Using heatmap to respresent correlation between features.
sns.heatmap(corr_matrix,annot=True,cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```

```python
#individual distribution of numeric columns
dataset.hist(figsize=(12, 10), color='skyblue', edgecolor='black')
plt.suptitle("Distribution of Variables")
plt.show()
```

```python
#Experience vs performance
sns.scatterplot(x='Experience (Years)',y='Performance Rating', hue='Department', data = dataset)
plt.title('Experience vs Performance Rating')
plt.show()
```

```python
# performance comparison
sns.boxplot(x='Department',y='Performance Rating',data = dataset)
plt.title("Performance Rating Comparison")
plt.show()
```
